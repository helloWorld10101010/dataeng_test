### Remove white spaces/na/empty values (Price Column)

```python
def remove_price_whitespaces(errorPrice):

    if type(errorPrice) == str:
        return errorPrice.lstrip()
    else:
        return errorPrice
```

### Main Transform method

```python
def remove_na_emptyString(NAEmptyValue):
    
    """    
    i) Split the name field into first_name, and last_name (split function " ")
    ii) Replace titles that are used before surnames or after last name (e.g Dr, Miss)
    iii) Remove any zeros prepended to the price field (Trim,clean)
    iv) Delete any rows which do not have a name (name field != "")
    v) Create a new field named above_100, which is true if the price is strictly greater than 100 (if price > 100, above_100 = 'TRUE', 'FALSE')
    """    

    if os.path.isfile(NAEmptyValue):

        dataset_raw = pd.read_csv(NAEmptyValue)
        dataset_dropna = dataset_raw.dropna(axis = 0, subset = ["name"])
        dataset_dropna_dropemty = dataset_dropna[dataset_dropna["name"] != ""]
        
        
        # clean name field
        
        dataset_dropna_dropemty["name"] = dataset_dropna_dropemty["name"].str.replace('Mr. ', '')
        dataset_dropna_dropemty["name"] = dataset_dropna_dropemty["name"].str.replace('Mrs. ', '')
        dataset_dropna_dropemty["name"] = dataset_dropna_dropemty["name"].str.replace('Dr. ', '')
        dataset_dropna_dropemty["name"] = dataset_dropna_dropemty["name"].str.replace('Ms. ', '')
        dataset_dropna_dropemty["name"] = dataset_dropna_dropemty["name"].str.replace('Miss. ', '')
        
        #get first name only
        
        dataset_dropna_dropemty["First_name"] = dataset_dropna_dropemty["name"].str.split(" ")
        
        dataset_dropna_dropemty['First_name'] = dataset_dropna_dropemty['First_name'].str[0]
        
        #get last name only
        
        dataset_dropna_dropemty["Last_name"] = dataset_dropna_dropemty["name"].str.split(" ")
        
        dataset_dropna_dropemty['Last_name'] = dataset_dropna_dropemty['Last_name'].str[1]
        
        #clean price(remove whitespaces)
    
        dataset_dropna_dropemty["Price"] = dataset_dropna_dropemty["price"].apply(lambda x: remove_price_whitespaces(x))
        
        #above_100(True/False)
        
        dataset_dropna_dropemty["Above_100"] = dataset_dropna_dropemty["price"].apply(lambda x: True if float(x) > 100 else False)
        
        #drop duplicate(working fields)
        dataset_dropna_dropemty = dataset_dropna_dropemty.drop(columns="name")
        dataset_dropna_dropemty = dataset_dropna_dropemty.drop(columns="price")

        return dataset_dropna_dropemty
    else:
        return NAEmptyValue
```
